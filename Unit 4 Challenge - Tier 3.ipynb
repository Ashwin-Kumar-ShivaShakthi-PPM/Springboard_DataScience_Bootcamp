**2.2. Cleaning the data**

You might find you need to transpose your dataframe, check out what its row indexes are, and reset the index. You  also might find you need to assign the values of the first row to your column headings  . (Hint: recall the .columns feature of DataFrames, as well as the iloc[] method).

```python
#Copy the native dataframe transposing to a new dataframe and Transpose the data structure to make the burroughs 
#become part of observational content

transposed_data = properties.T

#Review a sample of transposed data rows and columns
print('Before Reindexing: ', transposed_data.index)

#Initiate a reindex of the dataframe to ensure that a system generated numeric index is established instead
#of the column for burroughs. This would be required to access the first row in order to rename the  
transposed_data = transposed_data.reset_index()
print('After reindexing: ', transposed_data.index)

#Verify whether the reindex was successful
transposed_data.head()
```

**2.3. Cleaning the data (part 2)**

You might we have to **rename** a couple columns. How do you do this? The clue's pretty bold...

```python
#Rename the columns such that the sequential 0 thru n column names are renamed with values from the columns in row # 1
transposed_data.columns = transposed_data.iloc[0]

#Rename the first two columns to "London Burrough" and "ID" respectively
transposed_data = transposed_data.rename(columns = {'Unnamed: 0':'London_Borough', pd.NaT: 'ID'})

#Drop the first row as part of the cleanup
transposed_data.drop(0, inplace=True)

transposed_data.head()
```

**2.4.Transforming the data**

Remember what Wes McKinney said about tidy data? 

You might need to **melt** your DataFrame here. 

```python
#Use the melt function to convert the wide formatted datafram structure to long format for all the colums that are 
#dates
long_format_data = transposed_data.melt(id_vars=['London_Borough', 'ID'], var_name=['Time_Period'], value_name='Average_Price')
long_format_data.head()
```

Remember to make sure your column data types are all correct. Average prices, for example, should be floating point numbers... 

```python
#Convert the data type of Average Price from Object to float
print('Pre conversion Data Types: ', long_format_data.dtypes)
long_format_data['Average_Price'] = pd.to_numeric(long_format_data['Average_Price'])
print('Post conversion Data Types: ', long_format_data.dtypes)
```

**2.5. Cleaning the data (part 3)**

Do we have an equal number of observations in the ID, Average Price, Month, and London Borough columns? Remember that there are only 32 London Boroughs. How many entries do you have in that column? 

Check out the contents of the London Borough column, and if you find null values, get rid of them however you see fit. 

```python
'''
Cleansing will be performed through a series of steps
1. Remove the null values
2. Remove data about non-boroughs

Step 1
------
Run a test to see if there are null values. If there are dig deeper by the attributes where the null values are  
and identify a way to address those
'''
long_format_data.isna().sum()

#Use notna to extract the content that is not null and compare counts against isna
long_format_data.notna().sum()
long_format_data[long_format_data['ID'].notna()]
long_format_data[long_format_data['Average_Price'].notna()]

#Now apply the dropna to build a dataset that is free of null values that exists in the ID and Average Price columns
clean_dataset = long_format_data.dropna()

#Explore to determine that all nulls are indeed removed
clean_dataset[clean_dataset['ID'].isna()].sum()
clean_dataset[clean_dataset['Average_Price'].isna()].sum()
clean_dataset.notna().sum()
clean_dataset.count()
clean_dataset.shape

#Step 2
#------
#Remove data within the incoming dataset about non-boroughs

#Build a list of non-boroughs. 
nonBoroughs = ['Inner London', 'Outer London', 
               'NORTH EAST', 'NORTH WEST', 'YORKS & THE HUMBER', 
               'EAST MIDLANDS', 'WEST MIDLANDS',
              'EAST OF ENGLAND', 'LONDON', 'SOUTH EAST', 
              'SOUTH WEST', 'England']

#Check to see how many rows exist for non-burroughs
clean_dataset[clean_dataset.London_Borough.isin(nonBoroughs)]

#Check to see how many rows exist for actual burroughs
clean_dataset[~clean_dataset.London_Borough.isin(nonBoroughs)]

#There are 3804 rows for non-burroughs that need to be removed
#Subset for the content about actual burroughs alone and store it in the df, the cleanest dataset
df = clean_dataset[~clean_dataset.London_Borough.isin(nonBoroughs)]
df.shape
df.head()
df.dtypes

#This yields the content that would be graphically explored for further analysis
```

**2.6. Visualizing the data**

To visualize the data, why not subset on a particular London Borough? Maybe do a line plot of Month against Average Price?

**3. Modeling**

Consider creating a function that will calculate a ratio of house prices, comparing the price of a house in 2018 to the price in 1998.

Consider calling this function create_price_ratio.

You'd want this function to:
1. Take a filter of dfg, specifically where this filter constrains the London_Borough, as an argument. For example, one admissible argument should be: dfg[dfg['London_Borough']=='Camden'].
2. Get the Average Price for that Borough, for the years 1998 and 2018.
4. Calculate the ratio of the Average Price for 1998 divided by the Average Price for 2018.
5. Return that ratio.

Once you've written this function, you ultimately want to use it to iterate through all the unique London_Boroughs and work out the ratio capturing the difference of house prices between 1998 and 2018.

Bear in mind: you don't have to write a function like this if you don't want to. If you can solve the brief otherwise, then great! 

***Hint***: This section should test the skills you acquired in:
- Python Data Science Toolbox - Part One, all modules

```python
# Function for calculating the average increase in housing price for a specific burrough. Building the function 
# such that the average increase is calulated between the oldest year and the most current year. This is a diversion 
# from the ask about specific years (1998 and 2018) because the London datastore has been updated to include content
# for 2021
def create_price_ratio(bourough_data):
    oldest_year = float(bourough_data['Average_Price'][bourough_data['Year']== 1998])
    highest_year = float(bourough_data['Average_Price'][bourough_data['Year'] == 2021])
    ratio = [highest_year/oldest_year]
    return ratio

#Testing the function call
create_price_ratio(yearly_averages[yearly_averages['London_Borough']=='Camden'])

#Build a dictionary of average price increase for each bourough
final = {}

#Iterate through each bourough from the yearly_averages dataset that was built with grouping by year
for current_bourough in yearly_averages['London_Borough'].unique():
    # Subset the yearly_averages dataset to extract the current borough that was iterated 
    borough = yearly_averages[yearly_averages['London_Borough'] == current_bourough]
    
    # Make a new entry in the final dictionary for the current borough whose value is the result of calling 
    # create_price_ratio with the current borough data (subset of yearly_averages)
    final[current_bourough] = create_price_ratio(borough)
    
# Print the dictionary to validate the creation of price increase for each borough
print(final) 

#Transform the dictionary into a dataframe
#Transpose it to make the bouroughs become the context for the observations
#Reindex the dataset
Price_Ratios = pd.DataFrame(final).T.reset_index()

#Validate the successful completion of the above
Price_Ratios.head()

#Rename the first two columns as Bourough and 2021 to denote the bourough name and the current year
Price_Ratios.rename(columns={'index':'Borough', 0:'2021'}, inplace=True)

#Sort it in place by the descending order of the column titled 2021
Price_Ratios.sort_values(by='2021',ascending=False,inplace=True)

#Build a Bar graph to pictorially depict a comparison of average price inclrease across the top 15 bouroughs
top_15_bouroughs = Price_Ratios.head(15)
bar_graph = top_15_bouroughs[['Borough','2021']].plot(kind='bar')
bar_graph.set_xticklabels(top_15_bouroughs.Borough)
```
